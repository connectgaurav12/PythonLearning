```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
x
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("Title")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y)
axes2.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])
axes1.plot(x,z)
axes2.plot(x,y)
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color='blue',lw=2,ls='--')
axes[1].plot(x,z,color='red')
fig.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(5,5))
axes[0].plot(x,y,color='blue',lw=2,ls='--')
axes[1].plot(x,z,color='red')
fig.tight_layout()
```
