```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b')
plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Plot Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_title("LARGER Plot")
axes1.plot(x,y)

axes2.plot(y,x)
axes2.set_title("AXES 2 Title")
```

```python
##SUBPLOTS
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y)
#plt.tight_layout()\
axes[0].plot(x,y)
axes[1].plot(y,x)

axes[0].set_title("First AXES")
axes[1].set_title("Second Axes")
```

```python
type(axes)
```

**Figure Size and DPI**

```python
fig=plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_pic1.png',dpi=200)
```

```python
fig=plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.set_title("Title")
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc=0)
```

**Plot Appearance**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',lw=1,ls='-',marker='o',markersize='10',markerfacecolor='orange',markeredgecolor='red',markeredgewidth='3')
#ax.plot(y,x,color='blue',linewidth=5,linestyle='--')
```
